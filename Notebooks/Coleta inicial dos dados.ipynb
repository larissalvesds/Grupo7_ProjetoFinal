```python
import numpy as np
import pandas as pd

#  Lista as colunas numéricas de gasto
gastos = [
    "gastohospedagem",
    "gastoatrativos",
    "gastoalimentacao",
    "gastotransporte",
    "gastofatansias",
    "gastocompras",
    "gastoingressos",
    "gastooutros",
    "gastototal",
    "gastopercapto",
    "gastomediodiario",
    "gastomediodiariopercapto",
]

# 2) Colunas de FAIXA que já existem
faixas = [
    "faixagastohospedagem",
    "faixagastoatrativos",
    "faixagastoalimentacao",
    "faixagastotransporte",
    "faixagastofatansias",
    "faixagastocompras",
    "faixagastoingressos",
    "faixagastooutros",
    "faixagastototal",
    "faixagastopercapto",
    "faixagastomediodiario",
    "faixagastomediodiariopercapto",
]

#  Função para converter "R$ 1.234,56" -> 1234.56 e remover lixo
def to_number(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)
    s = str(x).strip()
    # remove tudo que não for dígito, vírgula, ponto, sinal
    s = pd.Series(s).str.replace(r"[^0-9,.\-]", "", regex=True).iloc[0]
    # padrão BR: troca milhar '.' por nada e decimal ',' por '.'
    s = s.replace(".", "").replace(",", ".")
    return pd.to_numeric(s, errors="coerce")

# Converte TODAS as colunas de gasto para numérico
for g in gastos:
    if g in df.columns:
        df[g] = df[g].apply(to_number)

# 5) Define bins e rótulos
bins = [-0.01, 0, 100, 500, 1000, float("inf")]
labels = ["Muito baixo (zero)", "Baixo", "Médio", "Alto", "Muito alto"]

# Preenche as colunas de faixa que JÁ existem (sem criar novas)
for g, f in zip(gastos, faixas):
    if g in df.columns and f in df.columns:
        df[f] = pd.cut(df[g], bins=bins, labels=labels, right=True)

# Checagens rápidas (opcional)
# se sobrou alguma linha que não virou número em cada coluna de gasto
na_report = {g: int(df[g].isna().sum()) for g in gastos if g in df.columns}
print("Valores não numéricos (após conversão) por coluna de gasto:", na_report)

# Frequência das faixas para uma coluna exemplo
if "faixagastototal" in df.columns:
    print(df["faixagastototal"].value_counts(dropna=False))
```

```python
# 📊 Análise exploratória dos gastos:
# Define as colunas de gastos

cols = [
    "gasto_hospedagem",
    "gasto_transporte",
    "gasto_alimentacao",
    "gasto_compras",
    "gasto_ingressos",
    "gasto_outros",
    "gasto_medio_diario",
    "gasto_total",
    "gasto_percapto"
]

# Mostra estatísticas descritivas detalhadas (inclui mediana, p90, p95, p99)
print(novo_df[cols].describe(percentiles=[.5, .9, .95, .99]).T)

# top valores
for c in cols:
    print("\nColuna:", c)
    print(novo_df[c].sort_values(ascending=False).head(5).to_list())


```

```python
"""Cortar outliers extremos (Winsorização)

Limitar os valores ao percentil 99 (ou 95 se quiser ser mais conservadora). Assim, ninguém “gasta mais que o top 1%”."""

for c in cols:
    limite = novo_df[c].quantile(0.99)
    novo_df[c] = novo_df[c].clip(upper=limite)
```

```python
"""3. Reportar média + mediana + p95/p99

Mediana = gasto típico da maioria.

Média sem outliers (winsorizada) = gasto médio mais realista.

p95/p99 = quanto gastam os maiores 5% ou 1%."""

comparacao = pd.DataFrame({
    "mediana": novo_df[cols].median(),
    "media_original": novo_df[cols].mean(),
    "media_sem_outliers": novo_df[cols].apply(lambda x: x.clip(upper=x.quantile(0.99)).mean())
})
print(comparacao)
```

```python

# calcula a média de cada coluna de gasto
medias = novo_df[cols].mean()

# define faixas: 0 é "Muito baixo", depois valores positivos
bins = [-0.01, 0, 100, 500, 1000, 2000, float("inf")]
labels = ["Muito baixo (0)", "Baixo", "Médio", "Alto", "Muito alto", "Extremo"]

# cria série categorizada
faixa_medias = pd.cut(medias, bins=bins, labels=labels, right=True, include_lowest=True)

# junta em DataFrame para visualizar
resumo = pd.DataFrame({"media": medias, "faixa": faixa_medias})
print(resumo)
```
